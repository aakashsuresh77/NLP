# 2) Use the above URL to scrape product links from product listing pages

```python
# Making a GET request
r = requests.get('https://www.amazon.in/')
total_content = r.content
# Parsing the HTML
soup = BeautifulSoup(total_content, 'html.parser')
```

```python
print(total_content)
# To view whether we have actually acquired amazon html data or amazon has blocked the request
```

```python
# find all the anchor tags with "href"
for link in soup.find_all('a'):
    if (link.get('href')) != None:
        if 'https://' in (link.get('href')):
            print(link.get('href'))
#Written a condition which checks and prints only https:// urls and not amazon local redirection url
```

# 4) Creation of CSV File
Construct csv file which includes following information (10 Marks)
                  a. Product title
                  b. Price
                  c. Model
                  d. Star rating

```python
r_csv = requests.get('https://www.flipkart.com/search?q=laptop&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&as-pos=1&as-type=HISTORY')
csv_content = r_csv.content
# Parsing the HTML
soup_csv = BeautifulSoup(csv_content, 'html.parser')
```

```python
print(soup_csv)
```

```python
import csv #Libarary to create and write data in CSV file
csv_file = open('product_detail.csv', 'w', encoding='utf-8') 
csv_writer = csv.writer(csv_file) 
csv_writer.writerow(['Product Title', 'Price', 'Model', 'Star rating'])
layer_one = soup_csv.find('div',class_="_1YokD2 _3Mn1Gg")
#print(layer_one)
k = 0
for link_csv in layer_one.find_all('div',class_="_1AtVbE col-12-12"):
    if k < 24:
        print("----------Next----------")
        #print(link)
        desc_model = link_csv.find('div',class_="_4rR01T")
        print(desc_model.text)
        desc_rating = link_csv.find('div',class_="_3LWZlK")
        if desc_rating != None:
            rating_ = desc_rating.text
        else:
            rating_ = "None"

        desc_price = link_csv.find('div',class_="_30jeq3 _1_WHN1")
        print(desc_price.text)
        csv_writer.writerow(['Laptop',desc_price.text,desc_model.text,rating_])
        k=k+1
    else:
        break
csv_file.close()
```

# 5) Create a corpus with 10 sample sentences. Perform Parts of Speech (POS) Tagging for the sentences in the corpus.

```python
import nltk

text = "I love eating apples. She goes to the store every day. He bought a new car last week. They are playing football in the park. She has been working on her project for two months. We need to finish the report today. The dog barked at the postman. The cat meowed at the mouse. I will be visiting my grandmother tomorrow. We can watch a movie after dinner."
#Text variable contains 10 samples
tokens = nltk.word_tokenize(text)
tagged = nltk.pos_tag(tokens)

print(tagged)
```

# 6) Find the Named Entity Recognition (NER) for the same corpus.

```python
!pip3 install spacy
```

```python
import spacy

nlp = spacy.load("en_core_web_sm")


doc = nlp(text)

for ent in doc.ents:
    print(ent.text, ent.label_)
```

# 7) Find the top 5 results for the following equation using Word2Vec Model "Einstein - Scientist + Picasso = ?"

```python
!pip3 install gensim
```

```python
import gensim.downloader as api
# This is a huge model (~1.6 gb) and it will take some time to load

```

```python
model = api.load('word2vec-google-news-300')
results = model.most_similar(positive=['Einstein', 'Picasso'], negative=['Scientist'], topn=5)

for result in results:
    print(result[0])
```
